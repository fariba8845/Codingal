```python
class canine:
  animal= "dog"
  def __init__(self, color, breed):
    self.color=color
    self.breed=breed
anna= canine("brown","pomeranian")
jake= canine("white","husky")

print("Anna is a",anna.animal)
print("She is a {} {}".format(anna.color, anna.breed))
print("Jake is a",jake.animal)
print("He is a {} {}".format(jake.color, jake.breed))
```
