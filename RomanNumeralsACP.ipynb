```python
class RomanNumerals:
    def __init__(self, num):
        self.num = num
    def transform(self):
        num = self.num
        val = [1000, 900, 500, 400, 100, 90, 50, 40, 10, 9, 5, 4, 1]
        syms = ["M", "CM", "D", "CD","C", "XC", "L", "XL","X", "IX", "V", "IV", "I"]
        result = ""
        for i in range(len(val)):
            while num >= val[i]:
                result += syms[i]
                num -= val[i]
        return result

inputt = int(input("Enter an integer: "))
romans = RomanNumerals(inputt)
print(romans.transform())
```
