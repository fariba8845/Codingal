```python
import random
import string

def random_password():
  password=[]
  num= int(input("Choose your password length: "))
  all_chars = string.ascii_uppercase + string.ascii_lowercase + string.digits
  for i in range(num):
    password.append(random.choice(all_chars))

  random.shuffle(password)
  final_pass= "".join(password)
  print("Your password is: ",final_pass)


random_password()
```
