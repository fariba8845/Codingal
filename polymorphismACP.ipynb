```python
from abc import ABC

class ferrari:
  def fuel_type(self):
    print("Ferraris use gasoline as fuel")
  def max_speed(self):
    print("Ferrari's maximum speed is 300 km/h")
class BMW:
  def fuel_type(self):
    print("BMWs use gasoline as fuel")
  def max_speed(self):
    print("BMW's maximum speed is 305 km/h")

obj_ferrari= ferrari()
obj_BMW= BMW()

for car in (obj_ferrari, obj_BMW):
  car.fuel_type()
  car.max_speed()
```
