```python

n= int(input("Enter any number: "))
i = 0
new_list= []
while i<=n:
  new_list.append(i)
  i+=1
odd= [x for x in new_list if x%2==1]
print("The list of all odd numbers:", odd)
print(odd)

list_2= ["apple","orange","mango","guava"]
list_final= []
for fruit in list_2:
    list_final.append(fruit.capitalize())
print("Original fruit list: ", list_2)
print("Capitalized fruit list: ", list_final)



```
